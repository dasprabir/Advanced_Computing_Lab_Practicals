```python
# ---------------------------------------------
# Breast Cancer Data Analysis using Python
# ---------------------------------------------
# Dataset: Breast Cancer Wisconsin (Diagnostic)
# Source: UCI ML Repository (via scikit-learn)
# ---------------------------------------------

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer


```

```python
# ---------------------------------------------
# 1. Load the Dataset
# ---------------------------------------------

# Load breast cancer dataset from sklearn
data = load_breast_cancer()

# Convert to Pandas DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add target variable (0 = malignant, 1 = benign)
df['diagnosis'] = data.target

# Replace numeric labels with meaningful names
df['diagnosis'] = df['diagnosis'].map({0: 'Malignant', 1: 'Benign'})

print("Dataset loaded successfully!")
print("Shape of dataset:", df.shape)
```

```python
# ---------------------------------------------
# 3. Descriptive Statistics
# ---------------------------------------------

# Overall statistics
print("\nOverall Statistical Summary:")
print(df.describe())

```

```python
# Mean, Variance, Standard Deviation using NumPy
mean_radius = np.mean(df['mean radius'])
var_radius = np.var(df['mean radius'])
std_radius = np.std(df['mean radius'])

print("\nStatistics for Mean Radius:")
print("Mean:", mean_radius)
print("Variance:", var_radius)
print("Standard Deviation:", std_radius)
```

```python
# ---------------------------------------------
# 4. Group-wise Statistical Analysis
# ---------------------------------------------

group_stats = df.groupby('diagnosis')['mean radius'].agg(
    ['mean', 'std', 'var', 'min', 'max']
)

print("\nGroup-wise Statistics (Mean Radius):")
print(group_stats)
```

```python
# ---------------------------------------------
# 5. Data Visualization
# ---------------------------------------------

# ---- Histogram ----
plt.figure()
plt.hist(df['mean radius'], bins=20)
plt.xlabel('Mean Radius')
plt.ylabel('Frequency')
plt.title('Distribution of Mean Radius')
plt.show()

# ---- Boxplot (Benign vs Malignant) ----
plt.figure()
df.boxplot(column='mean radius', by='diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Mean Radius')
plt.title('Mean Radius by Diagnosis')
plt.suptitle('')
plt.show()

# ---- Correlation Heatmap ----
correlation_matrix = df.drop('diagnosis', axis=1).corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.show()
```
