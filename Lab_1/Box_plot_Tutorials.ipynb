# Example 1: Comparing Two Groups (Male vs Female BP)

```python
import matplotlib.pyplot as plt

# Sample data: systolic blood pressure (mmHg)
male_bp = [110, 115, 120, 125, 130, 135, 140, 145]
female_bp = [105, 110, 115, 118, 122, 125, 130]

plt.figure()
plt.boxplot([male_bp, female_bp], labels=['Male', 'Female'])
plt.ylabel('Systolic Blood Pressure (mmHg)')
plt.title('Comparison of Blood Pressure by Gender')
plt.show()
```

# Example 2: Highlighting Outliers (Using IQR Rule)
## Teaching point

Points beyond whiskers = outliers

Whiskers follow 1.5 × IQR rule

```python
import matplotlib.pyplot as plt

# Data with outliers
data = [60, 62, 65, 68, 70, 72, 75, 78, 80, 120, 130]

plt.figure()
plt.boxplot(data, showfliers=True)
plt.ylabel('Measurement Value')
plt.title('Box Plot Showing Outliers')
plt.show()
```

# Example 3: Comparing Multiple Groups (Before vs After Treatment)
## Interpretation

Median decreases after treatment

Reduced variability → treatment effectiveness

```python
import matplotlib.pyplot as plt

# Blood glucose levels (mg/dL)
before_treatment = [180, 175, 170, 165, 160, 155, 150]
after_treatment = [140, 135, 130, 125, 120, 118, 115]

plt.figure()
plt.boxplot(
    [before_treatment, after_treatment],
    labels=['Before Treatment', 'After Treatment']
)
plt.ylabel('Blood Glucose (mg/dL)')
plt.title('Effect of Treatment on Blood Glucose')
plt.show()
```

# Example 4: Horizontal Box Plot (Better for Many Groups)
## Why horizontal?

Easier to read labels

Preferred when many categories exist

## Using the box plot, compare the central tendency and variability between the two groups.


```python
import matplotlib.pyplot as plt

group1 = [10, 12, 14, 15, 16, 18]
group2 = [20, 22, 25, 27, 28, 30]
group3 = [35, 38, 40, 42, 45, 48]

plt.figure()
plt.boxplot(
    [group1, group2, group3],
    vert=False,
    labels=['Group A', 'Group B', 'Group C']
)
plt.xlabel('Measurement Value')
plt.title('Horizontal Box Plot Comparison')
plt.show()
```
