```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/anonymized_marks.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
mean_marks = df['Marks'].mean()
median_marks = df['Marks'].median()
mode_marks = df['Marks'].mode()[0]
print("\nMean:", mean_marks)
print("Median:", median_marks)
print("Mode:", mode_marks)
```

```python
variance = df['Marks'].var()
std_dev = df['Marks'].std()
range_marks = df['Marks'].max() - df['Marks'].min()

print("\nVariance:", variance)
print("Standard Deviation:", std_dev)
print("Range:", range_marks)
```

```python
skewness = df['Marks'].skew()
kurtosis = df['Marks'].kurt()

print("\nSkewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
# Histogram
plt.figure()
plt.hist(df['Marks'], bins=10)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot
plt.figure()
plt.boxplot(df['Marks'])
plt.title("Box Plot of Marks")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure()
plt.plot(df['Marks'])
plt.title("Marks Trend")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```

```python
def assign_grade(marks):
    if marks >= 15:
        return 'A'
    elif marks >= 10:
        return 'B'
    elif marks >= 5:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Marks'].apply(assign_grade)

print("\nGrade Distribution:\n", df['Grade'].value_counts())
```

```python
# Bar Plot
plt.figure()
df['Grade'].value_counts().plot(kind='bar')
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()
```

```python
df.to_csv('processed_marks.csv', index=False)
print("\nProcessed data saved as 'processed_marks.csv'")
```
