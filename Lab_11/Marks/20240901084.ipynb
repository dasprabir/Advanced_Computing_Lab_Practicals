```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/anonymized_marks.csv')  # change name if needed
df.head()
```

```python
print(df.columns)
```

```python
col = "Marks"   # change if needed

data = df[col].dropna()

print("🔹 Highest (Max):", data.max())
print("🔹 Lowest (Min):", data.min())

print("🔹 Mean:", data.mean())
print("🔹 Median:", data.median())
print("🔹 Mode:", data.mode().tolist())

print("🔹 Standard Deviation:", data.std())
print("🔹 Variance:", data.var())
```

```python
stats = pd.DataFrame({
    "Min": [data.min()],
    "Max": [data.max()],
    "Mean": [data.mean()],
    "Median": [data.median()],
    "Mode": [data.mode().tolist()],
    "Std Dev": [data.std()],
    "Variance": [data.var()]
})

stats
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(data, bins=10, edgecolor='black')
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data, vert=True)
plt.title("Boxplot of Marks")
plt.ylabel("Marks")
plt.show()
```

```python
print("Skewness:", data.skew())
print("Kurtosis:", data.kurt())
```

```python
Q1 = data.quantile(0.25)
Q2 = data.quantile(0.50)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data < lower_bound) | (data > upper_bound)]

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:")
print(outliers.tolist())
```

```python
summary = pd.DataFrame({
    "Minimum": [data.min()],
    "Maximum": [data.max()],
    "Range": [data.max() - data.min()],
    "Mean": [data.mean()],
    "Median": [data.median()],
    "Mode": [data.mode().tolist()],
    "Standard Deviation": [data.std()],
    "Variance": [data.var()],
    "Q1": [Q1],
    "Q3": [Q3],
    "IQR": [IQR],
    "Skewness": [data.skew()],
    "Kurtosis": [data.kurt()],
    "Number of Outliers": [len(outliers)]
})

summary
```

```python
print("INTERPRETATION:")
print(f"1. Highest marks = {data.max()} and lowest marks = {data.min()}.")
print(f"2. Average marks (mean) = {data.mean():.2f}.")
print(f"3. Median marks = {data.median():.2f}.")
print(f"4. Mode = {data.mode().tolist()}.")
print(f"5. Standard deviation = {data.std():.2f}, showing the spread of marks.")
print(f"6. Variance = {data.var():.2f}.")
print(f"7. Skewness = {data.skew():.2f}.")
print(f"8. Kurtosis = {data.kurt():.2f}.")
print(f"9. Number of outliers = {len(outliers)}.")
```
