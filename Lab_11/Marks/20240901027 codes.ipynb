### Distribution of Marks: Visualizing the distribution of the 'Marks' column using a histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Marks'], kde=True, bins=15)
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Box Plot of Marks: Visualizing the distribution and spread of the 'Marks' column.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Marks'])
plt.title('Box Plot of Marks')
plt.ylabel('Marks')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Visualizing Standard Deviation on a Histogram: Plotting the mean and +/- 1 standard deviation.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Marks'], kde=True, bins=15)
plt.title('Distribution of Marks with Mean and Standard Deviation')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Calculate mean and standard deviation
mean_marks = df['Marks'].mean()
std_dev_marks = df['Marks'].std()

# Add vertical lines for mean and +/- 1 standard deviation
plt.axvline(mean_marks, color='red', linestyle='--', label=f'Mean: {mean_marks:.2f}')
plt.axvline(mean_marks - std_dev_marks, color='green', linestyle=':', label=f'-1 Std Dev: {mean_marks - std_dev_marks:.2f}')
plt.axvline(mean_marks + std_dev_marks, color='green', linestyle=':', label=f'+1 Std Dev: {mean_marks + std_dev_marks:.2f}')

plt.legend()
plt.show()
```

### Standard Deviation of Marks: Calculating and displaying the standard deviation of the 'Marks' column.

```python
std_dev_marks = df['Marks'].std()
print(f"Standard Deviation of Marks: {std_dev_marks:.2f}")
```
