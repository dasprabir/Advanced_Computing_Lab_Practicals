```python
import pandas as pd
import numpy as np

# ==============================
# 1. LOAD EXCEL FILE
# ==============================
df = pd.read_excel("/content/anonymized_marks.xlsx")  # change filename if needed

# If sheet name is specific:
# df = pd.read_excel("students.xlsx", sheet_name="Sheet1")

# ==============================
# 2. BASIC INFO
# ==============================
print("\n--- BASIC INFO ---")
print(df.info())

# ==============================
# 3. DESCRIPTIVE STATISTICS
# ==============================
print("\n--- DESCRIPTIVE STATISTICS ---")
print(df['Marks'].describe())

# ==============================
# 4. ADDITIONAL STATISTICS
# ==============================
print("\n--- ADDITIONAL METRICS ---")
print("Mean:", np.mean(df['Marks']))
print("Median:", np.median(df['Marks']))
print("Mode:", df['Marks'].mode().values)
print("Variance:", np.var(df['Marks'], ddof=1))
print("Standard Deviation:", np.std(df['Marks'], ddof=1))
print("Min:", np.min(df['Marks']))
print("Max:", np.max(df['Marks']))
print("Range:", np.max(df['Marks']) - np.min(df['Marks']))

# ==============================
# 5. QUARTILES
# ==============================
print("\n--- QUARTILES ---")
print("Q1:", df['Marks'].quantile(0.25))
print("Q2 (Median):", df['Marks'].quantile(0.50))
print("Q3:", df['Marks'].quantile(0.75))

# ==============================
# 6. SKEWNESS & KURTOSIS
# ==============================
print("\n--- DISTRIBUTION SHAPE ---")
print("Skewness:", df['Marks'].skew())
print("Kurtosis:", df['Marks'].kurt())

# ==============================
# 7. TOP & LOWEST
# ==============================
print("\n--- TOP PERFORMER ---")
print(df.loc[df['Marks'].idxmax()])

print("\n--- LOWEST PERFORMER ---")
print(df.loc[df['Marks'].idxmin()])

# ==============================
# 8. PASS/FAIL
# ==============================
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 5 else 'Fail')

print("\n--- PASS/FAIL COUNT ---")
print(df['Result'].value_counts())

# ==============================
# 9. SAVE OUTPUT
# ==============================
df.to_excel("Student_Report.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# LOAD EXCEL FILE
# ==============================
df = pd.read_excel("/content/anonymized_marks.xlsx")

marks = df['Marks']

# ==============================
# 1. HISTOGRAM (distribution)
# ==============================
plt.figure()
plt.hist(marks, bins=5)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# ==============================
# 2. BOXPLOT (outliers + spread)
# ==============================
plt.figure()
plt.boxplot(marks)
plt.title("Boxplot of Marks")
plt.ylabel("Marks")
plt.show()

# ==============================
# 3. BAR CHART (student vs marks)
# ==============================
plt.figure()
plt.bar(df['Student'], marks)
plt.title("Marks per Student")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.show()

# ==============================
# 4. LINE PLOT (trend)
# ==============================
plt.figure()
plt.plot(df['Student'], marks, marker='o')
plt.title("Marks Trend")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.show()

# ==============================
# 5. PIE CHART (pass/fail)
# ==============================
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 5 else 'Fail')

plt.figure()
df['Result'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pass vs Fail Distribution")
plt.ylabel("")
plt.show()

# ==============================
# 6. KDE PLOT (density curve)
# ==============================
plt.figure()
marks.plot(kind='kde')
plt.title("Density Plot of Marks")
plt.xlabel("Marks")
plt.show()

# ==============================
# 7. CUMULATIVE DISTRIBUTION
# ==============================
plt.figure()
marks.sort_values().reset_index(drop=True).plot()
plt.title("Cumulative Distribution of Marks")
plt.xlabel("Index")
plt.ylabel("Marks")
plt.show()

# ==============================
# 8. SCATTER PLOT (index vs marks)
# ==============================
plt.figure()
plt.scatter(range(len(marks)), marks)
plt.title("Scatter Plot of Marks")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```
