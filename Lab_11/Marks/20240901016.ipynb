```python
import pandas as pd

# Load dataset
df = pd.read_csv("anonymized_marks.csv")

# Sort by marks (highest first)
df = df.sort_values(by='Marks', ascending=False).reset_index(drop=True)

# Assign rank
df['Rank'] = df.index + 1

# Your custom grade scale
grades = ['A+', 'A', 'B+', 'B', 'C+', 'C', 'D', 'F']

# Assign grades (5 students per grade)
def assign_grade(rank):
    group = (rank - 1) // 5
    if group < len(grades):
        return grades[group]
    else:
        return 'F'  # remaining students stay F

df['Grade'] = df['Rank'].apply(assign_grade)

# Show result
print(df)

# Save file
df.to_csv("final_graded_students.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("anonymized_marks.csv")

# STEP 1: Sort and rank
df = df.sort_values(by='Marks', ascending=False).reset_index(drop=True)
df['Rank'] = df.index + 1

# STEP 2: Create grades (5 students per group)
grades = ['A+', 'A', 'B+', 'B', 'C+', 'C', 'D', 'F']

def assign_grade(rank):
    group = (rank - 1) // 5
    if group < len(grades):
        return grades[group]
    else:
        return 'F'

df['Grade'] = df['Rank'].apply(assign_grade)

# STEP 3: Count grades
grade_counts = df['Grade'].value_counts()

# STEP 4: Pie chart
plt.figure()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%')

plt.title("Grade Distribution")

plt.show()
```
