```python
# ===============================
# Step 1: Import Required Libraries
# ===============================
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# Step 2: Load the IRIS Dataset
# ===============================
# Update path if needed
df = pd.read_csv("/content/anonymized_marks.csv")

# Display first few rows
print("Dataset Preview:")
print(df.head())

# ===============================
# Step 3: Mean, Median, Mode

mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]

print("\nMean Values:")
print(mean_values)

print("\nMedian Values:")
print(median_values)

print("\nMode Values:")
print(mode_values)
```

```python
#Histogram Visualization
plt.figure()
plt.hist(df['Marks'], bins=10)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot
plt.figure()
plt.boxplot(df['Marks'])
plt.title("Boxplot of Marks")
plt.ylabel("Marks")
plt.show()
```

```python
#Scatter Plot
plt.figure()
plt.scatter(range(len(df)), df['Marks'])
plt.title("Scatter Plot of Students vs Marks")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure()
plt.fill_between(range(len(df)), df['Marks'])
plt.title("Area Plot of Marks")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```
