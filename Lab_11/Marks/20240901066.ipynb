```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
marks= pd.read_csv("/content/anonymized_marks.csv")
```

```python
marks.hist()
```

```python
sns.distplot(marks['Marks'])
```

```python
mean = marks['Marks'].mean()
var = marks['Marks'].var()
std = marks['Marks'].std()
mode=marks['Marks'].mode()
median=marks['Marks'].median()

print("\nStats for Marks:")
print("Mean:", mean)
print("Variance:", var)
print("Standard Deviation:", std)
print("Mode:", mode)
print("Median:", median)
```
