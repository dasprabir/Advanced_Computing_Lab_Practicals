```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the anonymized marks dataset if not already loaded
# Assuming marks_df is already loaded from previous steps. If not, uncomment the line below:
# marks_df = pd.read_csv('/content/anonymized_marks.csv')

# Define mark bins/categories
bins = [0, 5, 10, 15, 20]
labels = ['0-5', '6-10', '11-15', '16-20']

# Categorize 'Marks' into defined bins
marks_df['Mark_Category'] = pd.cut(marks_df['Marks'], bins=bins, labels=labels, right=True, include_lowest=True)

# Calculate the count of students in each category
category_counts = marks_df['Mark_Category'].value_counts().sort_index()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Anonymized Marks by Category')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution (mean and standard deviation)
mu = 0  # Mean
sigma = 1  # Standard deviation
sample_size = 1000 # Number of random samples

# Generate random numbers from a normal distribution
random_numbers = np.random.normal(mu, sigma, sample_size)

# Create a histogram of the random numbers
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=30, density=True, alpha=0.6, color='g', label='Histogram of Sampled Data')

# Plot the theoretical probability density function (PDF)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

plt.title('Normal Distribution: Histogram of Sampled Data vs. Theoretical PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the anonymized marks dataset if not already loaded
# Assuming marks_df is already loaded from previous steps. If not, uncomment the line below:
marks_df = pd.read_csv('/content/anonymized_marks.csv')

# Calculate the median of the 'Marks' column
median_marks = marks_df['Marks'].median()

# Create a histogram of the 'Marks' column
plt.figure(figsize=(10, 6))
plt.hist(marks_df['Marks'], bins=20, edgecolor='black', alpha=0.7, label='Marks Distribution')

# Add a vertical line for the median
plt.axvline(median_marks, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_marks:.2f}')

plt.title('Distribution of Anonymized Marks with Median')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()

print(f"The median of the 'Marks' column is: {median_marks:.2f}")
```
