```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/anonymized_marks.csv')

# Display the first few rows of the DataFrame to inspect the data
print("First 5 rows of the DataFrame:")
display(df.head())
```

Assuming there is a column named 'marks' in the dataset, let's calculate its mean, median, and mode. If your numerical column has a different name, please replace 'marks' in the code below.

```python
# Calculate the mean of the 'Marks' column
if 'Marks' in df.columns:
    mean_marks = df['Marks'].mean()
    print(f"Mean of 'Marks': {mean_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the median of the 'Marks' column
if 'Marks' in df.columns:
    median_marks = df['Marks'].median()
    print(f"Median of 'Marks': {median_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the mode(s) of the 'Marks' column
# Note: A column can have multiple modes.
if 'Marks' in df.columns:
    mode_marks = df['Marks'].mode()
    print(f"Mode(s) of 'Marks':\n{mode_marks}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the minimum of the 'Marks' column
if 'Marks' in df.columns:
    min_marks = df['Marks'].min()
    print(f"Lowest Mark: {min_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the maximum of the 'Marks' column
if 'Marks' in df.columns:
    max_marks = df['Marks'].max()
    print(f"Highest Mark: {max_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the standard deviation of the 'Marks' column
if 'Marks' in df.columns:
    std_marks = df['Marks'].std()
    print(f"Standard Deviation of 'Marks': {std_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
# Calculate the variance of the 'Marks' column
if 'Marks' in df.columns:
    var_marks = df['Marks'].var()
    print(f"Variance of 'Marks': {var_marks:.2f}")
else:
    print("Column 'Marks' not found. Please specify the correct column name.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Distribution of Marks: Histogram

```python
if 'Marks' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Marks'], kde=True, bins=10)
    plt.title('Distribution of Marks')
    plt.xlabel('Marks')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
else:
    print("Column 'Marks' not found. Cannot plot histogram.")
```

## Frequency Count of Each Mark

```python
import pandas as pd
df = pd.read_csv('/content/anonymized_marks.csv')

if 'Marks' in df.columns:
    mark_counts = df['Marks'].value_counts().sort_index()
    print("Frequency count of each unique mark:")
    display(mark_counts)
else:
    print("Column 'Marks' not found. Cannot calculate frequency counts.")
```

## Grade Distribution Table by Mark Ranges

```python
# Define mark ranges and labels
# Assuming marks are out of 20 for illustrative purposes. Adjust bins/labels as needed.
bins = [0, 5, 10, 15, 20]
labels = ['0-5', '5-10', '10-15', '15-20']

if 'Marks' in df.columns:
    # Create a new column 'Grade_Range' by categorizing 'Marks' into defined bins
    df['Grade_Range'] = pd.cut(df['Marks'], bins=bins, labels=labels, right=False)

    # Calculate the frequency of students in each grade range
    grade_distribution = df['Grade_Range'].value_counts().sort_index()

    print("Grade Distribution by Mark Ranges:")
    display(grade_distribution)
else:
    print("Column 'Marks' not found. Cannot create grade distribution table.")
```

## Grade Distribution: Bar Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns

if 'Grade_Range' in df.columns and not grade_distribution.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=grade_distribution.index, y=grade_distribution.values, hue=grade_distribution.index, palette='viridis', legend=False)
    plt.title('Grade Distribution by Mark Ranges')
    plt.xlabel('Mark Ranges')
    plt.ylabel('Number of Students')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
else:
    print("Grade distribution data not found. Please ensure 'Grade_Range' column is created and grade_distribution is calculated.")
```
