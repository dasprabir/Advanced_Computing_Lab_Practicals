```python
import pandas as pd

df = pd.read_csv('/content/anonymized_marks.csv')
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np

print('Statistical Summary of the DataFrame:')
display(df.describe())
```

### Distribution of Marks with Normal Curve Fit

Let's visualize the distribution of the 'Marks' column using a histogram and overlay a theoretical normal distribution curve. This will help us assess how closely the observed data follows a normal distribution.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Marks'], kde=True, stat='density', bins=15, color='skyblue', label='Actual Distribution')

# Calculate mean and standard deviation for 'Marks'
mu, std = df['Marks'].mean(), df['Marks'].std()

# Plot the PDF for a normal distribution with the calculated mean and std
x = np.linspace(df['Marks'].min(), df['Marks'].max(), 100)
plt.plot(x, norm.pdf(x, mu, std), color='red', linestyle='--', label=f'Normal Distribution (mu={mu:.2f}, std={std:.2f})')

plt.title('Distribution of Marks with Fitted Normal Curve')
plt.xlabel('Marks')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Box Plot of Marks

This box plot will show the median, quartiles, and potential outliers in the 'Marks' data.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Marks'], color='lightgreen')
plt.title('Box Plot of Marks')
plt.ylabel('Marks')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
