## Load Data and Inspect

```python
import pandas as pd

# Load the anonymized marks CSV file
try:
    df = pd.read_csv('/content/anonymized_marks (1).csv')
    print('CSV loaded successfully.')
    # Display the first 5 rows to understand the structure and column names
    display(df.head())
except FileNotFoundError:
    print('Error: The file \'/content/anonymized_marks (1).csv\' was not found. Please ensure it is uploaded correctly.')
except Exception as e:
    print(f'An error occurred while loading the CSV: {e}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

marks_column = 'Marks'

# Check if the 'Marks' column exists in the DataFrame
if marks_column in df.columns:
    # Create a histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(df[marks_column], bins=10, kde=True)
    plt.title(f'Distribution of {marks_column}')
    plt.xlabel(marks_column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
else:
    print(f"Error: Column '{marks_column}' not found in the DataFrame.")
```

```python
marks_column = 'Marks'

# Create a box plot
if marks_column in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[marks_column])
    plt.title(f'Box Plot of {marks_column}')
    plt.ylabel(marks_column)
    plt.grid(axis='y', alpha=0.75)
    plt.show()
else:
    print(f"Error: Column '{marks_column}' not found in the DataFrame.")
```

```python
import numpy as np
from scipy import stats

marks_column = 'Marks'

if marks_column in df.columns:
    # Extract the 'Marks' data
    marks_data = df[marks_column]

    # Calculate mean (average)
    mean_value = np.mean(marks_data)
    print(f"Mean (Average) of {marks_column}: {mean_value:.2f}")

    # Calculate median
    median_value = np.median(marks_data)
    print(f"Median of {marks_column}: {median_value:.2f}")

    # Calculate mode
    # stats.mode returns a ModeResult object, .mode[0] gets the value
    mode_result = stats.mode(marks_data, keepdims=True)
    mode_value = mode_result.mode[0]
    print(f"Mode of {marks_column}: {mode_value:.2f}")

    # Calculate standard deviation
    std_dev_value = np.std(marks_data)
    print(f"Standard Deviation of {marks_column}: {std_dev_value:.2f}")

    # Calculate highest mark
    highest_mark = np.max(marks_data)
    print(f"Highest {marks_column}: {highest_mark:.2f}")

    # Calculate lowest mark
    lowest_mark = np.min(marks_data)
    print(f"Lowest {marks_column}: {lowest_mark:.2f}")
else:
    print(f"Error: Column '{marks_column}' not found in the DataFrame.")
```
