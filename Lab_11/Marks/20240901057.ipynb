```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('/content/anonymized_marks.csv')
```

MEAN, MODE, **MEDIAN**

```python
marks = df['Marks']
mean_val = np.mean(marks)
median_val = np.median(marks)
mode_val = stats.mode(marks, keepdims=True)

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val.mode[0])
```

**AVERAGE**

```python
print("Average Marks:", mean_val)
```

**HIGHEST AND LOWEST**


```python
students = df.set_index('Student')['Marks'].to_dict()
max_student = max(students, key=students.get)
min_student = min(students, key=students.get)

print("Topper:", max_student, students[max_student])
print("Lowest:", min_student, students[min_student])
```

**HISTOGRAM**

```python


marks = list(students.values())

plt.hist(marks)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Marks Distribution")
plt.show()
```

STANDARD **DEV**

```python
import statistics
print("Std Dev:", statistics.stdev(marks))
```

**% WISE DISTRIBUTION**

```python
students = list(students.values())  # marks only

total = len(students)

# Initialize groups
ranges = {
    "0-5": 0,
    "5-10": 0,
    "10-15": 0,
    "15-20": 0
}

# Count students in each range
for m in students:
    if 0 <= m < 5:
        ranges["0-5"] += 1
    elif 5 <= m < 10:
        ranges["5-10"] += 1
    elif 10 <= m < 15:
        ranges["10-15"] += 1
    elif 15 <= m <= 20:
        ranges["15-20"] += 1

# Convert to percentage
for r in ranges:
    percent = (ranges[r] / total) * 100
    print(r, ":", round(percent, 2), "%")
```
