```python
# ===============================
# Step 1: Import Required Libraries
# ===============================
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# Step 2: Load the IRIS Dataset
# ===============================
# Update path if needed
df = pd.read_csv("IRIS.csv")

# Display first few rows
print("Dataset Preview:")
print(df.head())

# ===============================
# Step 3: Mean, Median, Mode
# ===============================
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]

print("\nMean Values:")
print(mean_values)

print("\nMedian Values:")
print(median_values)

print("\nMode Values:")
print(mode_values)

```

```python
# ===============================
# Step 4: Visualization 1 - Histogram
# ===============================
plt.figure()
plt.hist(df["sepal_length"], bins=20)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()

# ===============================
# Step 5: Visualization 2 - Boxplot
# ===============================
plt.figure()
df.drop(columns=["species"]).boxplot()
plt.title("Boxplot of Iris Features")
plt.show()

# ===============================
# Step 6: Visualization 3 - Scatter Plot
# ===============================
plt.figure()
plt.scatter(df["sepal_length"], df["petal_length"])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()

# ===============================
# Step 7: Visualization 4 - Bar Chart
# ===============================
plt.figure()
df["species"].value_counts().plot(kind="bar")
plt.title("Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
